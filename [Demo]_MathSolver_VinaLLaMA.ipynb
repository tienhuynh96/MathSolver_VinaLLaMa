## 3. Load pretrained LLM

```python
# Set model name
MODEL_NAME = "vilm/vinallama-7b-chat"

# Set config for Quantize
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,          # quantize 4 bit
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Load model
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    device_map="auto",
    trust_remote_code=True,
    quantization_config=bnb_config  # quantize config
)

# Create tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Print number trainable parameters
def print_trainable_parameters(model):
    """
    Prints the number of trainable parameters in the model.
    """
    trainable_params = 0
    all_param = 0

    for _, param in model.named_parameters():
        all_param += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    print(
        f"trainable params: {trainable_params} || all params: {all_param} || trainables%: {100 * trainable_params / all_param}"
    )
```

```python
# enables gradient checkpointing for the model.
# Gradient checkpointing is a technique used to reduce memory consumption during training
# It trades off increased computation for reduced memory usage by periodically saving intermediate activations and recomputing them during the backward pass.
# This is particularly useful when training very large models on GPUs with limited memory
model.gradient_checkpointing_enable()

# prepares the model for k-bit training.
# The function prepare_model_for_kbit_training is a utility that adjusts the model's layers and parameters to be compatible with low-precision training.
# This process is particularly useful for training large models on hardware with limited precision capabilities, like certain GPUs and specialized accelerators
model = prepare_model_for_kbit_training(model)
```

```python
# LoRa config
config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=[        # target module in model (check the model)
        "q_proj",
        "up_proj",
        "o_proj",
        "k_proj",
        "down_proj",
        "gate_proj",
        "v_proj"
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

# Add LoRa to model
model = get_peft_model(model, config)
# Print number of trainable parameters
print_trainable_parameters(model)
```

```python
# Create generate prompt function for instruction tuning
def generate_prompt(question, choices, explanation):
    return f"""
<|im_start|>system
Bạn là một chuyên gia về toán. Bạn sẽ nhận câu hỏi trắc nghiệm kèm theo các lựa chọn, hãy giải step by step nếu có và chọn phương án đúng.

<|im_start|>user
### Câu hỏi:
{question}
### Các lựa chọn:
{choices}
### Câu trả lời:

<|im_start|>assistant
{explanation}
""".strip()     # khi tạo prompt gắn thẳng lable (explanation), code của huggingface sẽ tự động tách ra để so sánh phân bố xác suất giữa lable and generation


# Create generate_and_tokenize_prompt function for create prompt and tokenizer them.
def generate_and_tokenize_prompt(question, choices, explanation):
    full_prompt = generate_prompt(question, choices, explanation)
    tokenized_full_prompt = tokenizer(
        full_prompt,
        padding=True,
        truncation=True
    )

    return tokenized_full_prompt
```

```python
# Create empty training sample variable
training_samples = []

# Loop through all training samples
for sample in tqdm(data['train']):
    # Get choices, explanation, question from each samples
    for quest in sample['problems']:
        choices = quest['choices']
        explanation = quest['explanation'].strip()
        question = quest['question']

        # Check if the question, choices, and explanation are not empty
        if explanation == '' or question == '' or choices == []:
            continue

        try:
            question = question.split('\n \n')[1].strip()
        except:
            continue

        # Join choices
        choices = '\n'.join(choices)
        # Generate and tokenizer prompt
        training_sample = generate_and_tokenize_prompt(
            question, choices, explanation
        )

        # append to training samples
        training_samples.append(training_sample)
```

```python
# Build dataset from list of training_samples
choices_data = Dataset.from_list(training_samples)
```

### 5.2. Training

```python
# Set training arguments
training_args = transformers.TrainingArguments(
      per_device_train_batch_size=1,
      gradient_accumulation_steps=4,
      num_train_epochs=1,
      learning_rate=2e-4,
      fp16=True,
      save_total_limit=3,
      logging_steps=1,
      output_dir="experiments",
      optim="paged_adamw_8bit",
      lr_scheduler_type="cosine",
      warmup_ratio=0.05,
)

# Set trainer
trainer = transformers.Trainer(
    model=model,
    train_dataset=choices_data,
    args=training_args,
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False)
)
# train model
model.config.use_cache = False
trainer.train()
```

### 5.3. Test prediction
